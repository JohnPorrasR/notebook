```python
data = pd.DataFrame(columns=('plato', 'personal', 'tiempo', 'precio', 'semana'))
```

```python
def crearLista(n, a, b):
    x = []
    for i in range(n):
        x.append(np.random.randint(a,b))
    return x
```

```python
data["plato"]    = crearLista(100, 10, 100)
data["personal"] = crearLista(100, 1, 5)
data["tiempo"]   = crearLista(100, 3, 10)
data["precio"]   = crearLista(100, 50, 100)
data["semana"]   = crearLista(100, 1, 7)
```

```python
np.savetxt("../datasets/practicas/don_panza.csv",   # Archivo de salida
           data,        # Trasponemos los datos se usa T
           fmt="%d",       # Usamos números enteros
           delimiter=","
           #newline="\n"
           )
```
