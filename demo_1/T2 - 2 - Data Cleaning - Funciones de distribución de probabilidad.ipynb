```python
import numpy as np
import matplotlib.pyplot as plt
```

## Distribución Uniforme

```python
a = 1
b = 100
n = 100000
data = np.random.uniform(a, b, n)
```

```python
%matplotlib inline
plt.hist(data)
```

## Distribución Normal

```python
data1 = np.random.randn(1000000)
```

```python
x = range(1,1000001)
plt.plot(x , data1)
```

```python
plt.hist(data1)
```

```python
plt.plot(x, sorted(data1))
```

```python
mu = 5.5
sd = 2.5
z  = np.random.randn(10000)
data2 = mu + sd * z
plt.hist(data2)
```
