## Generación aleatorios de números

```python
import numpy as np
```

```python
np.random.randint(1, 100)
```

```python
def randint_list(n, a, b):
    x = []
    for i in range(n):
        x.append(np.random.randint(a,b))
    return x
```

```python
randint_list(25, 1, 50)
```
