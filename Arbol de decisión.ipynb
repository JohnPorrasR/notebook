# Arbol de decisión

El algoritmo de árbol de decisión busca encontrar el rotulo de los datos por medio de preguntas:

Miremos un Ejemplo

![Arabol de decisión](./Arbol_de_decisión.png)

Para usar ese algoritmo (y otros) debemos de seguir siempre 4 pasos:
1. Impotar el algoritmo
2. Crear el clasificador
3. Entrenar el clasificador
4. Hacer predicción

```python
# 1 importamos las librerias
from sklearn import tree, datasets
import numpy as np
import matplotlib.pyplot as plt

# vamos a usar los datos de una planta
iris = datasets.load_iris()
# apenas vamos a usar dos caracteristicas o labels de estas plantas (ancho y largo de la hoja)
x = iris.data

# tenemos 3 tipos de planta rotulada (labels): 0, 1 y 2
y = iris.target
```

```python
# Generalmente todo nuestro set de satos no se debe usar para entrenar nuestro algoritmo
# la idea es poder tambien revisar la precisión del mismo
# entonces debemos de partir nuestro datos en datos de entrenamiento
# sklearn tiene este paso vital ya resulto para que nosotros lo hagamos de la forma mas sencilla posible

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0)
```

```python
# 2 creamos nuestro clasificador
clf = tree.DecisionTreeClassifier()
```

```python
# 3 Entrenamos nuestros clasificador usando los datos para entrenarlo por medio de la funciom fit
clf.fit(x_train, y_train)
```

```python
# 4 Hacemos predicciones con la funcion predict() que recibe como argumento las caracteristicas de prueba
pred = clf.predict(x_test)
```

```python
# vamos a importar una medición de la precisión de nuestro algoritmo
# osea que tambien esta haciendo nuestro trabajo nuestro clasificador
from sklearn.metrics import accuracy_score
acc = accuracy_score(pred, y_test)
print(acc)
```
