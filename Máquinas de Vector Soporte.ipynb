# Máquinas de Vector Soporte

## Ventajas
- Eficaz en espcios de alta dimensión
- Eficaz en casos en que el número es mayor que el número de muestras
- Es eficiente en la memoria

## Desventajas:
- Si el número de caracteristicas (features) es mucho mayor que el número de muestas, es probable que el método de un rendimiento pobre.

## Indicaciones

1. Import el algoritmo.
2. Crear el clasificador.
3. Entrenar el clasificador.

4. hacer predicciones.

```python
# 1 importamos las librerias
from sklearn import svm, datasets
import numpy as np
import matplotlib.pyplot as plt

# vamos a usar los datos de una planta
iris = datasets.load_iris()
# apenas vamos a usar dos caracteristicas o labels de estas plantas (ancho y largo de la hoja)
x = iris.data[:, :2]

# tenemos 3 tipos de planta rotulada (labels): 0, 1 y 2
y = iris.target
```

```python
# 2 creamos el clasificador
clf = svm.SVC(kernel='linear')
```

```python
# 3 entrenamos el algoritmo, cin fit
clf.fit(x, y)
```

```python
# no vamos a predecir nada en este ejemplo pero vamos a visualizar la información
plt.scatter(x[: ,0], x[:, 1], c=y)
plt.xlabel('Largo de la hoja')
plt.ylabel('Ancho de la hoja')
plt.title('SVM con kernel lineal')

plt.show()
```
