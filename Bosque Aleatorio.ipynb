# Método combinados - Bosque Aleatorio

Mejor que usar un solo algoritmo, sería poder usar algoritmos y sacar lo mejor de cada uno.

El algoritmo de bosque varios arboles de deciones (de ahi el nombre de bosque), y cada uno es entrenado en una porción aleatoria de los datos de entrada.

![Bosque Aleatorio](./Boque_aleatorio.png)

Esto se hace con el objetivo de mejorar la robustez del algoritmo

Existen muchos metodos de combinación de algoritmos y se conocen como metodos ensamblados o en ingles ** Ensemble methods ** como Adaboost, gradiente ... etc.

Para usar este algoritmo (y otro) debemos de seguir siempre los 4 pasos:

1. Impotar el algoritmo
2. Crear el clasificador
3. Entrenar el clasificador
4. Hacer predicción

```python
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# vamos a usar los datos de unas plantas
iris = datasets.load_iris()
# almacenaremos las caracteristicas de las plantas en la variable x
x = iris.data
# tenemos 3 tipos de planta rotulada (labels): 0, 1 y 2
y = iris.target
```

```python
# generalmente todo nuestro set de datos no se debe usar para entrenar nuestro algoritmo
# la idea es poder tambien revisar la precision del mismo
# entonces debemos de partir nuestros datos en datos de entrenamiento y dato de prueba
# sklearn tiene paso vital ya resuelto para que nosotros lo hagamos de la forma más sencilla posible
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0)
```

```python
# 2creamos nuestro clasificador donde n_estimators es el número de decision ue quereos usar
# para conformar el bosque
clf = RandomForestClassifier(n_estimators=100)
```

```python
# 3 entrenamos nuestro clasificador usando los datos para entrenarlo por medio de la funcion fit
clf.fit(x_train, y_train)
```

```python
# 4 hacemos predicciones con la funcion predict() que recibe como argumentos las caracteristicas de prueba
pred = clf.predict(x_test)
```

```python
# vamos a importar una medida de la precision de nuestro algoritmo
# osea que tan bien esta haciendo el trabajo nuestro clasificador
from sklearn.metrics import accuracy_score
acc = accuracy_score(pred, y_test)
print(acc)
```
