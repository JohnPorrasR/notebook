# Pandas

Es una libreria open source para análisis y procesamiento de datos, que demas facilita la creación de estructuras de datos de una forma muy sencilla.
Aprenderemos a leer archivos y transformarlos en estructuras de datos que nos van a facilitar mucho la vida a la hora de analizarlos.

## Archivos CSV (Local y web)

** CSV **: Comma separated values

```python
import pandas as pd
```

```python
raw_data = {
    'first_mane': ['John','Carmen','Camilla','Vladimir','Miguel'],
    'last_mane': ['Porras','Mendiola','Hurtado','Ramirez','Casana'],
    'age': [42, 52, 36, 24, 73],
    'presTestScore': [4, 24, 61, ".", "."],
    'postTestScore': ["25,000", "94,000", 27, 62, 70]
    
}
df = pd.DataFrame(raw_data)
df
```

### Guardar nuestro datos como csv

```python
df.to_csv('EjemploPadas.csv')
```

### Leer un archivo local CSV

```python
datos = pd.read_csv('EjemploPadas.csv')
datos
```

### Leer un CSV sin barra de titulos (Headers)

```python
datos = pd.read_csv('EjemploPadas.csv', header=None)
datos
```

```python
datos = pd.read_csv('EjemploPadas.csv', header=None, skiprows = 1)
datos
```

```python
datos = pd.read_csv('EjemploPadas.csv', index_col='UID', skiprows = 1, names=['UID','Nombre','Apellido','Edad','Puntaje 1','Puntaje 2'])
datos
```

### Leer CSV desde la web

```python
datos_web = pd.read_csv('http://winterolympicsmedals.com/medals.csv')
datos_web.head()
```
